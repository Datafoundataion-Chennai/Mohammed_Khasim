```python
#Stacked Bar Chart

import matplotlib.pyplot as plt
import numpy as np
months=["Jan","Feb","Mar","Apr","May","Jun"]
rent=[1200,1200,1200,1200,1250,1250]
food=[400,420, 450, 480, 500, 520]
entertainment = [150, 200, 220, 180, 250, 300]
transport = [100, 120, 130, 110, 140, 160]

rent=np.array(rent)
food=np.array(food)
entertainment=np.array(entertainment)
transport=np.array(transport)

plt.bar(months,rent,label="Rent",color=["red", "green", "blue", "orange", "purple", "cyan"])
plt.bar(months,food,bottom=rent,label="Food",color="green")
plt.bar(months,entertainment,bottom=rent+food,label="Entertainment",color="blue")
plt.bar(months,transport,bottom=rent+food+entertainment,label="Transport",color="purple")
plt.xlabel("Months")
plt.ylabel("Expenses")
plt.title("Monthly Expenses")
plt.legend()
plt.show()
```

```python
#SubPlots

import matplotlib.pyplot as plt
weeks = np.arange(1, 9)  
covid_cases = [500, 700, 1200, 2000, 3000, 4500, 6000, 8000]  
vaccination_rate = [5, 10, 20, 35, 50, 65, 75, 85]
fig,ax=plt.subplots(2,1,figsize=(10,6))
ax[0].plot(weeks,covid_cases,marker="o",color="blue",linestyle="--",label="Line Data")
ax[0].set_title("Covid Cases Over The Weeks")
ax[0].legend()
ax[0].grid(True)

ax[1].bar(weeks,vaccination_rate,color="green",label="Bar Data")
ax[1].set_title("Vaccination Dose Over The Week")
ax[1].legend()
ax[1].grid(True)

plt.tight_layout()
plt.show
```

```python
#Twin Y-axis

import numpy as np
import matplotlib.pyplot as plt

days = np.arange(1, 11)  
stock_price = [150, 152, 149, 155, 160, 162, 158, 165, 170, 175]  # In USD
trading_volume = [1.2, 1.4, 1.1, 1.8, 2.0, 2.5, 2.2, 2.7, 3.0, 3.5]

fig,ax1=plt.subplots()
ax1.set_xlabel("Days")
ax1.set_ylabel("Stock Price(USD)",color="blue")
ax1.plot(days,stock_price,marker="o",linestyle="--",color="green",label="Stock Price")
ax1.tick_params(axis='y',color="cyan")

ax2=ax1.twinx()
ax2.set_ylabel("Trading Volume(MS)",color="red")
ax2.plot(days,trading_volume,marker="s",linestyle="--",color="grey",label="Trading Volume")
ax2.tick_params(axis='y',color="violet")
plt.title("Stock Price Trading Over Ten Days")
ax1.grid(True)
plt.show()
```
